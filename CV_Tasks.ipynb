```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read img and convert color
img = cv2.imread('images.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# build a transformation matrix
matrix = [[1, 0, 0], #x 
          [0, 1, 0]] #y
t = np.float32(matrix)
# get the sizes
h, w = img.shape[:2]
# transform
img = cv2.warpAffine(img, t, (w, h))
# plot
fig, ax = plt.subplots(1, figsize=(12,8))
ax.axis('on')   
plt.imshow(img)
```

```python
# Change location
#25% of x-direction (25% of 200=50)
matrix = [[1, 0, 50], #x-direction is positive towards right direction
          [0, 1, 0]]
```

```python
t = np.float32(matrix)
# get the sizes
h, w = img.shape[:2]
# transform
img_x_shift = cv2.warpAffine(img, t, (w, h))
# plot
fig, ax = plt.subplots(1, figsize=(12,8))
ax.axis('on')   
plt.imshow(img_x_shift)
plt.savefig("x_direction.jpg", dpi= 100, bbox_inches = 'tight')
```

```python
# read img and convert color
img = cv2.imread('images.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# build a transformation matrix
matrix = [[1, 0, 0], #x 
          [0, 1, 0]] #y
t = np.float32(matrix)
# get the sizes
h, w = img.shape[:2]
# transform
img = cv2.warpAffine(img, t, (w, h))
# plot
fig, ax = plt.subplots(1, figsize=(12,8))
ax.axis('on')   
plt.imshow(img)
#25% of y-direction (25% of 250=62.5)
y_matrix = [[1, 0, 0],
            [0, 1, 62.5]] #y-direction is positive downwards
t = np.float32(y_matrix)
# get the sizes
h, w = img.shape[:2]
# transform
img_y_shift = cv2.warpAffine(img, t, (w, h))
# plot
fig, ax = plt.subplots(1, figsize=(12,8))
ax.axis('on')   
plt.imshow(img_y_shift)
plt.savefig("y_direction.jpg", dpi= 100, bbox_inches = 'tight')
```

```python
import cv2

img=cv2.imread("images.jpg")

# theta is positive, therefore rotation is counter-clock-wise 
out=cv2.transpose(img)
out=cv2.flip(out,flipCode=0)
rotate_first = cv2.imwrite("rotated_counterclockwise.jpg", out)
cv2.imshow("image", out)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
img=cv2.imread("images.jpg")

# theta is negative, therefore rotation is clock-wise 
out=cv2.transpose(img)
out=cv2.flip(out,flipCode=1)
rotate_first = cv2.imwrite("rotated_clockwise.jpg", out)
cv2.imshow("image", out)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
